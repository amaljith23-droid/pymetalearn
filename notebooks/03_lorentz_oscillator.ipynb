## 3.1 Physical Picture: Bound Electrons

### Drude Model Limitation

The Drude model treats electrons as **free** (no restoring force). This works for:
- Conduction electrons in metals
- Below interband transition energies

### Lorentz Model Extension

For **bound electrons** (in atoms, interband transitions), we add a **restoring force**:
- Electron is bound to the nucleus like a mass on a spring
- Natural oscillation frequency $\omega_0$
- Examples: core electrons, dielectrics, absorption bands in metals

```python
# Visualize bound electron model
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Drude: free electron
ax1 = axes[0]
ax1.plot(0, 0, 'ro', markersize=30, label='Nucleus (+)')
for angle in np.linspace(0, 2*np.pi, 5)[:-1]:
    x = 0.8 * np.cos(angle)
    y = 0.8 * np.sin(angle)
    ax1.plot(x, y, 'bo', markersize=10)
    ax1.annotate('', xy=(x + 0.3*np.cos(angle), y + 0.3*np.sin(angle)), 
                xytext=(x, y), arrowprops=dict(arrowstyle='->', color='blue'))
ax1.set_xlim(-1.5, 1.5)
ax1.set_ylim(-1.5, 1.5)
ax1.set_aspect('equal')
ax1.set_title('Drude Model\n(Free electrons, no restoring force)', fontsize=12)
ax1.axis('off')

# Lorentz: bound electron
ax2 = axes[1]
ax2.plot(0, 0, 'ro', markersize=30, label='Nucleus (+)')
# Draw spring
t = np.linspace(0, 0.7, 50)
spring_y = 0.1 * np.sin(20 * t)
ax2.plot(t, spring_y, 'k-', linewidth=2)
ax2.plot(0.8, 0, 'bo', markersize=15, label='Electron (-)')

# Restoring force arrow
ax2.annotate('', xy=(0.5, 0), xytext=(0.8, 0),
            arrowprops=dict(arrowstyle='->', color='green', lw=2))
ax2.text(0.65, 0.15, r'$F_{spring} = -k x$', fontsize=11, color='green')

ax2.set_xlim(-0.5, 1.5)
ax2.set_ylim(-0.8, 0.8)
ax2.set_aspect('equal')
ax2.set_title('Lorentz Model\n(Bound electron with spring restoring force)', fontsize=12)
ax2.axis('off')

plt.tight_layout()
plt.show()
```

## 3.2 Derivation: Equation of Motion for Bound Electrons

### Forces on the Electron

1. **Driving force** from electric field: $F_E = -eE$
2. **Damping force** (collisions/radiation): $F_d = -m_e\gamma \dot{x}$
3. **Restoring force** (NEW!): $F_s = -m_e\omega_0^2 x$

where $\omega_0$ is the natural resonance frequency.

```python
# Symbolic derivation
t = symbols('t', real=True)
x = Function('x')(t)
m, e_sym, E_0, omega, gamma, omega_0 = symbols('m_e e E_0 omega gamma omega_0', positive=True, real=True)

display(Markdown("### Step 1: Write Newton's second law"))
display(Math(r"m_e \frac{d^2x}{dt^2} = F_{field} + F_{damping} + F_{restoring}"))

display(Math(r"m_e \frac{d^2x}{dt^2} = -eE_0 e^{-i\omega t} - m_e\gamma\frac{dx}{dt} - m_e\omega_0^2 x"))

display(Markdown("Rearranging:"))
display(Math(r"\boxed{m_e \frac{d^2x}{dt^2} + m_e\gamma\frac{dx}{dt} + m_e\omega_0^2 x = -eE_0 e^{-i\omega t}}"))

display(Markdown("This is the **driven damped harmonic oscillator** equation."))
```

```python
display(Markdown("### Step 2: Assume harmonic solution"))
display(Math(r"x(t) = x_0 e^{-i\omega t}"))

display(Markdown("Derivatives:"))
display(Math(r"\dot{x} = -i\omega x_0 e^{-i\omega t}"))
display(Math(r"\ddot{x} = -\omega^2 x_0 e^{-i\omega t}"))

display(Markdown("### Step 3: Substitute into equation of motion"))
display(Math(r"m_e(-\omega^2) x_0 + m_e\gamma(-i\omega) x_0 + m_e\omega_0^2 x_0 = -eE_0"))

display(Markdown("Where we cancelled the common $e^{-i\omega t}$ factor."))

display(Markdown("### Step 4: Solve for displacement amplitude"))
display(Math(r"m_e x_0 (\omega_0^2 - \omega^2 - i\gamma\omega) = -eE_0"))

display(Math(r"\boxed{x_0 = \frac{eE_0}{m_e(\omega_0^2 - \omega^2 - i\gamma\omega)}}"))
```

## 3.3 Derivation: The Lorentz Dielectric Function

Following the same steps as Drude:
- Polarization: $P = -n_e e x_0$
- Permittivity: $\varepsilon_r = 1 + P/(\varepsilon_0 E)$

```python
display(Markdown("### Step 5: Calculate polarization"))
display(Math(r"P = -n_e e \cdot x_0 = -n_e e \cdot \frac{eE_0}{m_e(\omega_0^2 - \omega^2 - i\gamma\omega)}"))
display(Math(r"P = \frac{-n_e e^2 E_0}{m_e(\omega_0^2 - \omega^2 - i\gamma\omega)}"))

display(Markdown("### Step 6: Calculate permittivity"))
display(Math(r"\varepsilon_r = 1 + \frac{P}{\varepsilon_0 E_0}"))
display(Math(r"= 1 - \frac{n_e e^2}{\varepsilon_0 m_e (\omega_0^2 - \omega^2 - i\gamma\omega)}"))

display(Markdown("Define the **oscillator strength** plasma frequency:"))
display(Math(r"\omega_p^2 = \frac{n_e e^2}{\varepsilon_0 m_e}"))

display(Markdown("**The Lorentz Permittivity:**"))
display(Math(r"\boxed{\varepsilon_r(\omega) = 1 + \frac{\omega_p^2}{\omega_0^2 - \omega^2 - i\gamma\omega}}"))
```

```python
display(Markdown("### Comparison: Drude vs Lorentz"))

display(Math(r"\text{Drude (free electrons):} \quad \varepsilon_r = 1 - \frac{\omega_p^2}{\omega^2 + i\gamma\omega}"))
display(Math(r"\text{Lorentz (bound electrons):} \quad \varepsilon_r = 1 + \frac{\omega_p^2}{\omega_0^2 - \omega^2 - i\gamma\omega}"))

display(Markdown("**Note**: Drude is the $\\omega_0 \\to 0$ limit of Lorentz!"))
display(Markdown("(No restoring force means no natural frequency.)"))
```

## 3.4 Derivation: Real and Imaginary Parts

Let's separate $\varepsilon_r = \varepsilon_1 + i\varepsilon_2$

```python
display(Markdown("### Step 7: Rationalize the denominator"))

display(Markdown("Let $D = \\omega_0^2 - \\omega^2 - i\\gamma\\omega$"))
display(Math(r"D^* = \omega_0^2 - \omega^2 + i\gamma\omega"))

display(Math(r"\varepsilon_r = 1 + \frac{\omega_p^2}{D} \cdot \frac{D^*}{D^*}"))
display(Math(r"= 1 + \frac{\omega_p^2 (\omega_0^2 - \omega^2 + i\gamma\omega)}{(\omega_0^2 - \omega^2)^2 + (\gamma\omega)^2}"))

display(Markdown("### Step 8: Extract real and imaginary parts"))
display(Math(r"\boxed{\varepsilon_1 = 1 + \frac{\omega_p^2 (\omega_0^2 - \omega^2)}{(\omega_0^2 - \omega^2)^2 + \gamma^2\omega^2}}"))
display(Math(r"\boxed{\varepsilon_2 = \frac{\omega_p^2 \gamma \omega}{(\omega_0^2 - \omega^2)^2 + \gamma^2\omega^2}}"))
```

```python
# Lorentz permittivity function
def lorentz_permittivity(omega, omega_p, omega_0, gamma):
    """Complex Lorentz permittivity."""
    return 1 + omega_p**2 / (omega_0**2 - omega**2 - 1j*gamma*omega)

# Normalized frequency range
omega_norm = np.linspace(0.01, 3, 1000)
omega_0_norm = 1.0
omega_p_norm = 0.5
gamma_norm = 0.1

eps = lorentz_permittivity(omega_norm, omega_p_norm, omega_0_norm, gamma_norm)

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Real part
ax1 = axes[0]
ax1.plot(omega_norm, eps.real, 'b-', linewidth=2)
ax1.axhline(y=1, color='gray', linestyle='--', label='Îµ = 1')
ax1.axhline(y=0, color='k', linestyle='-', linewidth=0.5)
ax1.axvline(x=omega_0_norm, color='red', linestyle=':', linewidth=2, label=r'$\omega_0$')
ax1.fill_between(omega_norm, eps.real, 0, where=(eps.real < 0), 
                 alpha=0.3, color='red', label='Îµâ‚ < 0')

ax1.set_xlabel(r'$\omega / \omega_0$', fontsize=14)
ax1.set_ylabel(r'$\varepsilon_1$', fontsize=14)
ax1.set_title('Real Part: Anomalous Dispersion Near Resonance', fontsize=13)
ax1.legend(loc='upper right')
ax1.grid(True, alpha=0.3)
ax1.set_ylim(-2, 4)

# Imaginary part (absorption)
ax2 = axes[1]
ax2.plot(omega_norm, eps.imag, 'r-', linewidth=2)
ax2.axvline(x=omega_0_norm, color='red', linestyle=':', linewidth=2, label=r'$\omega_0$')
ax2.fill_between(omega_norm, eps.imag, 0, alpha=0.3, color='orange', label='Absorption')

ax2.set_xlabel(r'$\omega / \omega_0$', fontsize=14)
ax2.set_ylabel(r'$\varepsilon_2$', fontsize=14)
ax2.set_title('Imaginary Part: Absorption Peak at Resonance', fontsize=13)
ax2.legend()
ax2.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print("Key features:")
print("â€¢ Îµâ‚ can become NEGATIVE just above resonance!")
print("â€¢ Îµâ‚‚ peaks at the resonance frequency")
print("â€¢ This is the basis for engineered metamaterial resonances")
```

## 3.5 Key Physics: When Does Îµâ‚ < 0?

From our derived formula:
$$\varepsilon_1 = 1 + \frac{\omega_p^2 (\omega_0^2 - \omega^2)}{(\omega_0^2 - \omega^2)^2 + \gamma^2\omega^2}$$

For $\varepsilon_1 < 0$, we need the fraction to be negative and larger than 1.

```python
display(Markdown("### Condition for Îµâ‚ < 0"))

display(Markdown("For $\\omega > \\omega_0$: numerator $(\\omega_0^2 - \\omega^2) < 0$ â†’ fraction is negative"))

display(Markdown("The condition $\\varepsilon_1 < 0$ requires:"))
display(Math(r"\frac{\omega_p^2 (\omega^2 - \omega_0^2)}{(\omega_0^2 - \omega^2)^2 + \gamma^2\omega^2} > 1"))

display(Markdown("**In the low-damping limit ($\\gamma \\ll \\omega_0$), near resonance:**"))
display(Math(r"\varepsilon_1 \approx 1 - \frac{\omega_p^2}{\omega^2 - \omega_0^2}"))

display(Markdown("This is negative for $\\omega_0 < \\omega < \\sqrt{\\omega_0^2 + \\omega_p^2}$"))
```

```python
# Show negative epsilon region
fig, ax = plt.subplots(figsize=(12, 6))

omega_p_values = [0.3, 0.5, 0.8, 1.0]
gamma_val = 0.05  # Low damping

for omega_p_val in omega_p_values:
    eps = lorentz_permittivity(omega_norm, omega_p_val, 1.0, gamma_val)
    ax.plot(omega_norm, eps.real, linewidth=2, label=f'$\omega_p/\omega_0$ = {omega_p_val}')

ax.axhline(y=0, color='k', linestyle='-', linewidth=1)
ax.axvline(x=1.0, color='gray', linestyle=':', linewidth=2, label=r'$\omega_0$')

ax.set_xlabel(r'$\omega / \omega_0$', fontsize=14)
ax.set_ylabel(r'$\varepsilon_1$', fontsize=14)
ax.set_title('Effect of Oscillator Strength: Stronger $\omega_p$ â†’ Wider Îµ < 0 Region', fontsize=14)
ax.legend()
ax.grid(True, alpha=0.3)
ax.set_ylim(-5, 5)
ax.set_xlim(0.5, 2)

plt.tight_layout()
plt.show()

print("Larger oscillator strength Ï‰â‚š â†’ wider frequency band with Îµ < 0")
```

## 3.6 The Drude-Lorentz Model for Real Metals

Real metals have **both**:
1. Free electrons (Drude contribution)
2. Bound electrons / interband transitions (Lorentz contribution)

$$\varepsilon_r = \underbrace{1 - \frac{\omega_{p,free}^2}{\omega^2 + i\gamma_{free}\omega}}_{\text{Drude (free)}} + \sum_j \underbrace{\frac{f_j \omega_{p,j}^2}{\omega_{0,j}^2 - \omega^2 - i\gamma_j\omega}}_{\text{Lorentz (bound)}}$$

```python
def drude_lorentz(omega, omega_p_drude, gamma_drude, lorentz_params):
    """
    Combined Drude-Lorentz permittivity.
    
    lorentz_params: list of (omega_0, omega_p, gamma, f) tuples
    """
    # Drude term
    eps = 1 - omega_p_drude**2 / (omega**2 + 1j*gamma_drude*omega)
    
    # Lorentz terms
    for omega_0, omega_p, gamma, f in lorentz_params:
        eps += f * omega_p**2 / (omega_0**2 - omega**2 - 1j*gamma*omega)
    
    return eps

# Gold model parameters (simplified)
omega_p_Au = 1.37e16  # rad/s
gamma_Au = 4.05e13    # rad/s

# Add interband transition at ~2.4 eV (green/blue absorption)
eV_to_rad = 1.519e15  # rad/s per eV
lorentz_Au = [
    (2.4 * eV_to_rad, 0.3 * omega_p_Au, 0.2 * eV_to_rad, 0.8),
]

# Wavelength range
wavelengths = np.linspace(300, 1000, 500)  # nm
omega = 2 * np.pi * c / (wavelengths * 1e-9)

# Drude only
eps_drude = 1 - omega_p_Au**2 / (omega**2 + 1j*gamma_Au*omega)

# Drude-Lorentz
eps_DL = drude_lorentz(omega, omega_p_Au, gamma_Au, lorentz_Au)

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

ax1 = axes[0]
ax1.plot(wavelengths, eps_drude.real, 'b--', linewidth=2, label='Drude only')
ax1.plot(wavelengths, eps_DL.real, 'b-', linewidth=2, label='Drude-Lorentz')
ax1.axhline(y=0, color='k', linestyle='-', linewidth=0.5)
ax1.set_xlabel('Wavelength (nm)', fontsize=14)
ax1.set_ylabel(r'$\varepsilon_1$', fontsize=14)
ax1.set_title('Gold: Real Part of Permittivity', fontsize=14)
ax1.legend()
ax1.grid(True, alpha=0.3)

ax2 = axes[1]
ax2.plot(wavelengths, eps_drude.imag, 'r--', linewidth=2, label='Drude only')
ax2.plot(wavelengths, eps_DL.imag, 'r-', linewidth=2, label='Drude-Lorentz')
ax2.set_xlabel('Wavelength (nm)', fontsize=14)
ax2.set_ylabel(r'$\varepsilon_2$', fontsize=14)
ax2.set_title('Gold: Imaginary Part (Absorption)', fontsize=14)
ax2.legend()
ax2.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print("The interband transition at ~500 nm causes gold's yellow color!")
print("(Blue light is absorbed, yellow/red is reflected)")
```

## 3.7 Metamaterial Application: Engineered Resonances

**Key insight**: The Lorentz model shows that resonances create Îµ < 0!

We can **engineer** artificial resonators (SRRs, nanoparticles) to create:
- Resonances at any desired frequency
- Îµ < 0 above the resonance
- Î¼ < 0 using magnetic resonators (Chapter 6)

```python
# Design an artificial resonator for 10 GHz
f_target = 10e9  # 10 GHz
omega_target = 2 * np.pi * f_target

# Frequency range
freq = np.linspace(1, 20, 500)  # GHz
omega = 2 * np.pi * freq * 1e9

omega_0 = omega_target
omega_p = 0.3 * omega_0  # Moderate oscillator strength
gamma = 0.05 * omega_0   # Low damping

eps = lorentz_permittivity(omega, omega_p, omega_0, gamma)

fig, ax = plt.subplots(figsize=(12, 6))

ax.plot(freq, eps.real, 'b-', linewidth=2, label=r'$\varepsilon_1$')
ax.plot(freq, eps.imag, 'r-', linewidth=2, label=r'$\varepsilon_2$')
ax.axhline(y=0, color='k', linestyle='-', linewidth=0.5)
ax.axvline(x=10, color='gray', linestyle=':', linewidth=2, label='Resonance (10 GHz)')
ax.fill_between(freq, eps.real, 0, where=(eps.real < 0), 
                alpha=0.3, color='red', label='Îµ < 0 region')

ax.set_xlabel('Frequency (GHz)', fontsize=14)
ax.set_ylabel('Permittivity', fontsize=14)
ax.set_title('Engineered Resonator: Îµ < 0 Above Resonance', fontsize=14)
ax.legend(loc='upper right')
ax.grid(True, alpha=0.3)
ax.set_ylim(-5, 5)

plt.tight_layout()
plt.show()

print("By engineering resonators with specific Ï‰â‚€, we can create Îµ < 0 at any frequency!")
print("This is the foundation of electric metamaterials.")
```

## 3.8 Summary: Complete Derivation

### Derivation Chain

| Step | Equation |
|------|----------|
| 1. Equation of motion | $m_e \ddot{x} + m_e\gamma\dot{x} + m_e\omega_0^2 x = -eE$ |
| 2. Harmonic solution | $x_0 = eE_0/[m_e(\omega_0^2 - \omega^2 - i\gamma\omega)]$ |
| 3. Polarization | $P = -n_e e x_0$ |
| 4. Permittivity | $\varepsilon_r = 1 + \omega_p^2/(\omega_0^2 - \omega^2 - i\gamma\omega)$ |

### Key Physics

- **Below resonance** ($\omega < \omega_0$): $\varepsilon_1 > 1$ (normal dispersion)
- **At resonance** ($\omega = \omega_0$): Strong absorption peak
- **Above resonance** ($\omega > \omega_0$): $\varepsilon_1 < 1$, can be negative!

### Metamaterial Insight

By engineering artificial resonators with desired $\omega_0$, we can create $\varepsilon < 0$ at any frequency!
