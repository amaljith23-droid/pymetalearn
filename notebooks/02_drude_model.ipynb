## 2.1 Derivation: Equation of Motion for Free Electrons

### Physical Picture

In a metal, valence electrons are **delocalized** and can move freely through the lattice of positive ions.

When an electric field $\mathbf{E}$ is applied:
1. Electrons experience a force $\mathbf{F} = -e\mathbf{E}$
2. Electrons collide with lattice ions/impurities (damping)
3. Average time between collisions = $\tau$ (relaxation time)

### Newton's Second Law

For a single electron at position $\mathbf{r}$:

$$m_e \frac{d^2\mathbf{r}}{dt^2} = \mathbf{F}_{field} + \mathbf{F}_{damping}$$

````python
# Define symbols
t = symbols('t', real=True)
x = Function('x')(t)
m, e_charge, E_0, omega, gamma, tau = symbols('m_e e E_0 omega gamma tau', positive=True, real=True)

display(Markdown("### Step 1: Write the forces"))

display(Markdown("**Electric force** (from applied AC field $E = E_0 e^{-i\omega t}$):"))
display(Math(r"F_{field} = -eE = -eE_0 e^{-i\omega t}"))

display(Markdown("**Damping force** (friction from collisions, proportional to velocity):"))
display(Math(r"F_{damping} = -m_e \gamma \frac{dx}{dt}"))

display(Markdown("where $\\gamma = 1/\\tau$ is the damping rate (collision frequency)."))
````

````python
display(Markdown("### Step 2: Write the equation of motion"))

display(Math(r"m_e \frac{d^2x}{dt^2} = -eE_0 e^{-i\omega t} - m_e \gamma \frac{dx}{dt}"))

display(Markdown("Rearranging:"))
display(Math(r"\boxed{m_e \frac{d^2x}{dt^2} + m_e \gamma \frac{dx}{dt} = -eE_0 e^{-i\omega t}}"))

display(Markdown("This is a **driven damped harmonic oscillator** equation (without restoring force)."))
````

---

## 2.2 Derivation: Solving for Electron Displacement

### Assume Harmonic Solution

Since the driving field oscillates as $e^{-i\omega t}$, we assume:
$$x(t) = x_0 e^{-i\omega t}$$

where $x_0$ is the (complex) amplitude of oscillation.

````python
x_0 = symbols('x_0', complex=True)

display(Markdown("### Step 3: Substitute the assumed solution"))

display(Math(r"x(t) = x_0 e^{-i\omega t}"))
display(Math(r"\frac{dx}{dt} = -i\omega x_0 e^{-i\omega t}"))
display(Math(r"\frac{d^2x}{dt^2} = (-i\omega)^2 x_0 e^{-i\omega t} = -\omega^2 x_0 e^{-i\omega t}"))

display(Markdown("### Step 4: Substitute into equation of motion"))
display(Math(r"m_e(-\omega^2 x_0 e^{-i\omega t}) + m_e \gamma(-i\omega x_0 e^{-i\omega t}) = -eE_0 e^{-i\omega t}"))

display(Markdown("### Step 5: Cancel the $e^{-i\omega t}$ factors (they appear everywhere)"))
display(Math(r"m_e(-\omega^2 x_0) + m_e \gamma(-i\omega x_0) = -eE_0"))
display(Math(r"-m_e x_0 (\omega^2 + i\gamma\omega) = -eE_0"))
````

````python
display(Markdown("### Step 6: Solve for $x_0$"))

display(Math(r"x_0 = \frac{eE_0}{m_e(\omega^2 + i\gamma\omega)}"))

display(Markdown("Factor out $\\omega$ from denominator:"))
display(Math(r"\boxed{x_0 = \frac{eE_0}{m_e \omega(\omega + i\gamma)}}"))

display(Markdown("**Key insight**: The displacement is **complex** â†’ there is a phase lag between E and x."))
````

## 2.3 Derivation: The Drude Dielectric Function

### From Displacement to Polarization

The **polarization** $\mathbf{P}$ is the dipole moment per unit volume.

Each electron contributes a dipole moment $-ex$, so for $n_e$ electrons per unit volume:

$$P = -n_e e x$$

````python
n_e, P, D = symbols('n_e P D', real=True)
eps_0, eps_r = symbols('varepsilon_0 varepsilon_r', positive=True)

display(Markdown("### Step 7: Calculate polarization"))

display(Math(r"P = -n_e e \cdot x_0 = -n_e e \cdot \frac{eE_0}{m_e(\omega^2 + i\gamma\omega)}"))
display(Math(r"P = \frac{-n_e e^2 E_0}{m_e(\omega^2 + i\gamma\omega)}"))

display(Markdown("### Step 8: Relate to electric displacement"))
display(Markdown("From Maxwell's equations, the electric displacement is:"))
display(Math(r"D = \varepsilon_0 E + P = \varepsilon_0 \varepsilon_r E"))

display(Markdown("Therefore:"))
display(Math(r"\varepsilon_0 \varepsilon_r E = \varepsilon_0 E + P"))
display(Math(r"\varepsilon_r = 1 + \frac{P}{\varepsilon_0 E}"))
````

````python
display(Markdown("### Step 9: Substitute P to get Îµ_r"))

display(Math(r"\varepsilon_r = 1 + \frac{1}{\varepsilon_0 E_0} \cdot \frac{-n_e e^2 E_0}{m_e(\omega^2 + i\gamma\omega)}"))

display(Math(r"\varepsilon_r = 1 - \frac{n_e e^2}{\varepsilon_0 m_e(\omega^2 + i\gamma\omega)}"))

display(Markdown("### Step 10: Define the plasma frequency"))
display(Math(r"\boxed{\omega_p^2 \equiv \frac{n_e e^2}{\varepsilon_0 m_e}}"))

display(Markdown("**The Drude dielectric function:**"))
display(Math(r"\boxed{\varepsilon_r(\omega) = 1 - \frac{\omega_p^2}{\omega^2 + i\gamma\omega}}"))
````

## 2.4 Physical Meaning of the Plasma Frequency

The plasma frequency $\omega_p$ is the **natural oscillation frequency** of the electron gas.

### Physical Picture

If electrons are displaced from equilibrium:
1. A restoring electric field arises (from charge separation)
2. Electrons oscillate back and forth at frequency $\omega_p$
3. These are called **plasma oscillations** or **plasmons**

````python
# Calculate plasma frequencies for common metals
metals = {
    'Silver (Ag)': 5.86e28,
    'Gold (Au)': 5.90e28,
    'Copper (Cu)': 8.47e28,
    'Aluminum (Al)': 18.1e28,
}

print("Plasma Frequencies of Common Metals")
print("=" * 70)
print(f"{'Metal':<20} {'n_e (mâ»Â³)':<15} {'Ï‰â‚š (rad/s)':<15} {'fâ‚š (PHz)':<10} {'Î»â‚š (nm)':<10}")
print("-" * 70)

for metal, n_e_val in metals.items():
    omega_p = np.sqrt(n_e_val * e**2 / (epsilon_0 * m_e))
    f_p = omega_p / (2 * np.pi)
    lambda_p = c / f_p * 1e9  # nm
    print(f"{metal:<20} {n_e_val:<15.2e} {omega_p:<15.2e} {f_p/1e15:<10.2f} {lambda_p:<10.1f}")

print("\nðŸ’¡ Î»â‚š is in the UV range â†’ metals are reflective in the visible!")
````

## 2.5 Derivation: Real and Imaginary Parts of Îµ

The dielectric function is complex: $\varepsilon_r = \varepsilon_1 + i\varepsilon_2$

Let's separate the real and imaginary parts algebraically.

````python
omega_p_sym = symbols('omega_p', positive=True, real=True)

display(Markdown("### Step 11: Separate real and imaginary parts"))

display(Markdown("Starting from:"))
display(Math(r"\varepsilon_r = 1 - \frac{\omega_p^2}{\omega^2 + i\gamma\omega}"))

display(Markdown("**Multiply numerator and denominator by the complex conjugate of the denominator:**"))
display(Math(r"\varepsilon_r = 1 - \frac{\omega_p^2}{\omega^2 + i\gamma\omega} \cdot \frac{\omega^2 - i\gamma\omega}{\omega^2 - i\gamma\omega}"))

display(Math(r"= 1 - \frac{\omega_p^2(\omega^2 - i\gamma\omega)}{(\omega^2)^2 + (\gamma\omega)^2}"))

display(Math(r"= 1 - \frac{\omega_p^2(\omega^2 - i\gamma\omega)}{\omega^2(\omega^2 + \gamma^2)}"))
````

````python
display(Markdown("### Step 12: Extract real and imaginary parts"))

display(Math(r"\varepsilon_r = 1 - \frac{\omega_p^2 \omega^2}{\omega^2(\omega^2 + \gamma^2)} + i\frac{\omega_p^2 \gamma\omega}{\omega^2(\omega^2 + \gamma^2)}"))

display(Math(r"= \left(1 - \frac{\omega_p^2}{\omega^2 + \gamma^2}\right) + i\frac{\omega_p^2 \gamma}{\omega(\omega^2 + \gamma^2)}"))

display(Markdown("**Final result:**"))
display(Math(r"\boxed{\varepsilon_1 = 1 - \frac{\omega_p^2}{\omega^2 + \gamma^2}}"))
display(Math(r"\boxed{\varepsilon_2 = \frac{\omega_p^2 \gamma}{\omega(\omega^2 + \gamma^2)}}"))
````

````python
display(Markdown("### Key Limiting Cases"))

display(Markdown("**Low damping limit ($\\gamma \\ll \\omega$):**"))
display(Math(r"\varepsilon_1 \approx 1 - \frac{\omega_p^2}{\omega^2}"))
display(Math(r"\varepsilon_2 \approx \frac{\omega_p^2 \gamma}{\omega^3}"))

display(Markdown("**At the plasma frequency ($\\omega = \\omega_p$, $\\gamma \\ll \\omega_p$):**"))
display(Math(r"\varepsilon_1 \approx 1 - 1 = 0"))

display(Markdown("**Below plasma frequency ($\\omega < \\omega_p$):**"))
display(Math(r"\varepsilon_1 < 0 \quad \text{(NEGATIVE!)}"))
````

````python
# Visualize Drude permittivity
def drude_permittivity(omega, omega_p, gamma):
    """Calculate complex Drude permittivity."""
    return 1 - omega_p**2 / (omega**2 + 1j * gamma * omega)

# Silver parameters
n_e_Ag = 5.86e28
omega_p_Ag = np.sqrt(n_e_Ag * e**2 / (epsilon_0 * m_e))
gamma_Ag = 3.23e13  # Typical value for silver

omega_ratio = np.linspace(0.01, 3, 500)
omega = omega_ratio * omega_p_Ag
eps = drude_permittivity(omega, omega_p_Ag, gamma_Ag)

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Real part
ax1 = axes[0]
ax1.plot(omega_ratio, eps.real, 'b-', linewidth=2, label=r'$\varepsilon_1$')
ax1.axhline(y=0, color='k', linestyle='--', linewidth=1)
ax1.axvline(x=1, color='r', linestyle=':', linewidth=2, label=r'$\omega = \omega_p$')
ax1.fill_between(omega_ratio, eps.real, 0, where=(eps.real < 0), 
                 alpha=0.3, color='red', label='Îµâ‚ < 0')
ax1.set_xlabel(r'$\omega / \omega_p$', fontsize=14)
ax1.set_ylabel(r'$\varepsilon_1$ (real part)', fontsize=14)
ax1.set_title('Real Part: Negative Below Ï‰â‚š!', fontsize=14)
ax1.legend()
ax1.grid(True, alpha=0.3)
ax1.set_xlim(0, 3)
ax1.set_ylim(-20, 5)

# Imaginary part
ax2 = axes[1]
ax2.plot(omega_ratio, eps.imag, 'r-', linewidth=2, label=r'$\varepsilon_2$')
ax2.axvline(x=1, color='r', linestyle=':', linewidth=2)
ax2.set_xlabel(r'$\omega / \omega_p$', fontsize=14)
ax2.set_ylabel(r'$\varepsilon_2$ (imaginary part)', fontsize=14)
ax2.set_title('Imaginary Part: Absorption', fontsize=14)
ax2.legend()
ax2.grid(True, alpha=0.3)
ax2.set_xlim(0, 3)

plt.tight_layout()
plt.show()
````

## 2.6 Derivation: Optical Constants from Permittivity

The complex refractive index is:
$$\tilde{n} = n + i\kappa = \sqrt{\varepsilon_r}$$

where:
- **n**: refractive index (phase velocity = c/n)
- **Îº**: extinction coefficient (absorption)

````python
display(Markdown("### Step 13: Derive n and Îº from Îµâ‚ and Îµâ‚‚"))

display(Markdown("From $\\tilde{n}^2 = \\varepsilon_r$:"))
display(Math(r"(n + i\kappa)^2 = \varepsilon_1 + i\varepsilon_2"))
display(Math(r"n^2 - \kappa^2 + 2in\kappa = \varepsilon_1 + i\varepsilon_2"))

display(Markdown("Matching real and imaginary parts:"))
display(Math(r"n^2 - \kappa^2 = \varepsilon_1"))
display(Math(r"2n\kappa = \varepsilon_2"))

display(Markdown("Solving these simultaneous equations (algebra omitted):"))
display(Math(r"\boxed{n = \sqrt{\frac{\sqrt{\varepsilon_1^2 + \varepsilon_2^2} + \varepsilon_1}{2}}}"))
display(Math(r"\boxed{\kappa = \sqrt{\frac{\sqrt{\varepsilon_1^2 + \varepsilon_2^2} - \varepsilon_1}{2}}}"))
````

````python
display(Markdown("### Step 14: Derive the skin depth"))

display(Markdown("A wave propagating in the z-direction:"))
display(Math(r"E = E_0 e^{i(kz - \omega t)} = E_0 e^{i(\tilde{n}k_0 z - \omega t)}"))
display(Math(r"= E_0 e^{i((n + i\kappa)k_0 z - \omega t)}"))
display(Math(r"= E_0 e^{-\kappa k_0 z} e^{i(nk_0 z - \omega t)}"))

display(Markdown("The amplitude decays as $e^{-\\kappa k_0 z} = e^{-z/\\delta}$ where:"))
display(Math(r"\boxed{\delta = \frac{1}{\kappa k_0} = \frac{\lambda_0}{2\pi\kappa}}"))

display(Markdown("This is the **skin depth** - the distance over which the field decays to 1/e."))
````

````python
# Calculate optical properties for silver
wavelengths = np.linspace(200, 1000, 500) * 1e-9
omega_wl = 2 * np.pi * c / wavelengths
eps_Ag = drude_permittivity(omega_wl, omega_p_Ag, gamma_Ag)

# Calculate n and kappa using the derived formulas
eps1 = eps_Ag.real
eps2 = eps_Ag.imag
n = np.sqrt((np.sqrt(eps1**2 + eps2**2) + eps1) / 2)
kappa = np.sqrt((np.sqrt(eps1**2 + eps2**2) - eps1) / 2)

# Skin depth
skin_depth = wavelengths / (2 * np.pi * kappa) * 1e9  # nm

# Reflectivity (derived in next chapter, using Fresnel)
R = np.abs((1 - n - 1j*kappa) / (1 + n + 1j*kappa))**2

fig, axes = plt.subplots(1, 3, figsize=(15, 4))

ax1 = axes[0]
ax1.plot(wavelengths*1e9, n, 'b-', linewidth=2, label='n')
ax1.plot(wavelengths*1e9, kappa, 'r-', linewidth=2, label='Îº')
ax1.set_xlabel('Wavelength (nm)', fontsize=12)
ax1.set_ylabel('Optical Constants', fontsize=12)
ax1.set_title('Silver: n and Îº (Drude Model)', fontsize=12)
ax1.legend()
ax1.grid(True, alpha=0.3)

ax2 = axes[1]
ax2.plot(wavelengths*1e9, skin_depth, 'g-', linewidth=2)
ax2.set_xlabel('Wavelength (nm)', fontsize=12)
ax2.set_ylabel('Skin Depth Î´ (nm)', fontsize=12)
ax2.set_title('Skin Depth in Silver', fontsize=12)
ax2.set_ylim(0, 50)
ax2.grid(True, alpha=0.3)

ax3 = axes[2]
ax3.plot(wavelengths*1e9, R*100, 'm-', linewidth=2)
ax3.set_xlabel('Wavelength (nm)', fontsize=12)
ax3.set_ylabel('Reflectivity (%)', fontsize=12)
ax3.set_title('Silver Reflectivity', fontsize=12)
ax3.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
````

## 2.7 Summary: Complete Derivation Chain

```
Equation of Motion â†’ Displacement xâ‚€ â†’ Polarization P â†’ Permittivity Îµ
       â†“                    â†“               â†“              â†“
  F = ma - friction    Harmonic sol.    P = -n_eÂ·eÂ·x    Îµ = 1 + P/(Îµâ‚€E)
```

### Key Results

| Quantity | Formula |
|----------|--------|
| Plasma frequency | $\omega_p = \sqrt{n_e e^2 / \varepsilon_0 m_e}$ |
| Drude permittivity | $\varepsilon_r = 1 - \omega_p^2 / (\omega^2 + i\gamma\omega)$ |
| Real part | $\varepsilon_1 = 1 - \omega_p^2 / (\omega^2 + \gamma^2)$ |
| Imaginary part | $\varepsilon_2 = \omega_p^2\gamma / [\omega(\omega^2 + \gamma^2)]$ |
| Skin depth | $\delta = \lambda / (2\pi\kappa)$ |

### Why This Matters for Metamaterials

**Below $\omega_p$: $\varepsilon_1 < 0$** â†’ First ingredient for negative index!
